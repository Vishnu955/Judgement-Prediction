```python
# Configure the Gemini API key
#os.environ['GEMINI_API_KEY'] = 'YOUR_API_KEY'
genai.configure(api_key=os.environ['GEMINI_API_KEY'])

# Function to classify a single text input
def classify_text(input_text):
    prompt = (
        "You are a legal assistant. Classify the following court judgment into one of these categories:\n" "- Appeal allowed\n" "- Appeal dismissed\n" "- Petition dismissed\n" "- Petition allowed\n\n"
        f"Here is the court judgment:\n{input_text}\n\n" "Please respond with the exact category name from the list above.")
    model = genai.GenerativeModel("gemini-1.5-flash")
    response = model.generate_content(prompt)
    # Extract the model's classification from the response
    classification = response.text.strip()
    return classification 

```

```python
# Apply GEMINI classification to the 'Judgement' column and store predictions
data['Predicted_Class'] = data['Judgement'].apply(lambda x: classify_text(x) if pd.notnull(x) else "Unknown")
```
